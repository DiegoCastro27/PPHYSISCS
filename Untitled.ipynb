```python
import random
import math
import matplotlib.pyplot as plt

# Parmetros de la caja
ANCHO = 100
ALTO = 80
NUM_PARTICULAS = 20
RADIO_PARTICULA = 2  # Radio de cada partcula

# Funcin para verificar que las partculas no se solapen
def verificar_colision(particula, particulas_existentes):
    for p in particulas_existentes:
        distancia = math.sqrt((particula[0] - p[0]) ** 2 + (particula[1] - p[1]) ** 2)
        if distancia < 2 * RADIO_PARTICULA:
            return False
    return True

# Funcin para generar las partculas
def generar_particulas():
    particulas = []
    while len(particulas) < NUM_PARTICULAS:
        # Generar una posicin aleatoria dentro de la caja
        x = random.uniform(RADIO_PARTICULA, ANCHO - RADIO_PARTICULA)
        y = random.uniform(RADIO_PARTICULA, ALTO - RADIO_PARTICULA)
        
        # Verificar que no colisione con las partculas existentes
        if verificar_colision((x, y), particulas):
            particulas.append((x, y))
    return particulas

# Generar las partculas
particulas = generar_particulas()

# Visualizar la caja y las partculas
plt.figure(figsize=(6, 5))
plt.xlim(0, ANCHO)
plt.ylim(0, ALTO)
plt.gca().set_aspect('equal', adjustable='box')

# Dibujar las partculas
for particula in particulas:
    circle = plt.Circle(particula, RADIO_PARTICULA, color='blue', alpha=0.6)
    plt.gca().add_artist(circle)

# Configuracin de la caja
plt.gca().add_patch(plt.Rectangle((0, 0), ANCHO, ALTO, linewidth=2, edgecolor='black', facecolor='none'))

plt.title("Caja con 20 partculas que no se tocan")
plt.show()


import numpy as np
import matplotlib.pyplot as plt

# Parmetros de la caja
ANCHO = 100
ALTO = 80
NUM_PARTICULAS = 20
RADIO_PARTICULA = 2  # Radio de cada partcula

# Funcin para generar las partculas equidistantes
def generar_particulas_equidistantes():
    filas = 4
    columnas = 5
    particulas = []
    
    # Calcular el espacio entre partculas para que no se solapen
    espacio_x = (ANCHO - 2 * RADIO_PARTICULA) / (columnas - 1)
    espacio_y = (ALTO - 2 * RADIO_PARTICULA) / (filas - 1)
    
    for i in range(filas):
        for j in range(columnas):
            x = RADIO_PARTICULA + j * espacio_x
            y = RADIO_PARTICULA + i * espacio_y
            particulas.append((x, y))
    
    return particulas

# Generar las partculas
particulas = generar_particulas_equidistantes()

# Visualizar la caja y las partculas
plt.figure(figsize=(6, 5))
plt.xlim(0, ANCHO)
plt.ylim(0, ALTO)
plt.gca().set_aspect('equal', adjustable='box')

# Dibujar las partculas
for particula in particulas:
    circle = plt.Circle(particula, RADIO_PARTICULA, color='red', alpha=0.6)
    plt.gca().add_artist(circle)

# Configuracin de la caja
plt.gca().add_patch(plt.Rectangle((0, 0), ANCHO, ALTO, linewidth=2, edgecolor='black', facecolor='none'))

plt.title("Caja con 20 partculas equidistantes")
plt.show()
```
